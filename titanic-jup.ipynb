```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df['satisfaction_level'] = df['pclass'].map({1: 'high', 2:'medium', 3:'low'})
df.head()
```

```python
df.isna().sum()
```

```python
df = df.drop('deck', axis=1)
df['age']= df['age'].fillna(df['age'].median())

df.head()
```

```python
df.isna().sum()
```

```python
df.dropna()
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split

x= df[['satisfaction_level']]
y = df[['alive']]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
print(x_train.head())
type(x_train)
```

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(categories=[['low', 'medium', 'high']])

x_train_encoded = oe.fit_transform(x_train)
x_test_encoded = oe.transform(x_test)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train.values.ravel())
y_test_encoded = le.fit_transform(y_test.values.ravel())
```
